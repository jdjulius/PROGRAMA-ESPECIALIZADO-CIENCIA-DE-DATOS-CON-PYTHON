# Diccionarios

Es una colección de elementos, donde cada uno tiene una llave "key" y un valor "value". 

Los diccionarios se pueden crear con llave {} separando con una coma cada par key:value.

```python
dic1 = {} # un diccionario vacío
dic1
```

```python
dicVehiculo = {
  "marca": "Ford",
  "modelo": "Mustang",
  "año": 1964
}
dicVehiculo
```

```python
dicVehiculo['año'] # obteniendo el valor de un "Key"
```

```python
dicVehiculo['marca']
```

```python
# Si el key al que accedemos no existe, se añade automáticamente.
dicVehiculo['nro asientos'] = [5, 4, 4, 3]
dicVehiculo
```

```python
dicVehiculo['nro asientos'] = [10, 20]  # actualizando / modificando  un valor asociado a un key
dicVehiculo
```

## Recorriendo los diccionarios

Podemos recorrer de manera similar que a las Tuplas y Listas.

```python
l = list() # lista vacía

for k in dicVehiculo: #imprimirá los "Key"
    print(k)
    l.append(k)

print(l)  # lista con los Key del diccionario
```

```python
l = list()

for v in dicVehiculo:  # key
    print(dicVehiculo[v]) #imprimirá los valores del Diccionario
    
    l.append(dicVehiculo[v])
l
```

```python
# Imprime los key y value del diccionario
# diccionario key:valor ---> elemento de un diccionario

for k, v in dicVehiculo.items():  
    print("Key: " + str(k) + " y el Valor: " + str(v))
```

### pop(<key>[,<default>])

Busca y elimina la key y su valor asociado.

```python
dic2.pop('a')  # key
print(dic2) # muestra lo que quedó del diccionario
```

```python
dic2.pop('z', 'c') # Si no encuentra, mostrará la lista sin cambios
print(dic2)       
```

```python
dic2.pop('b', -1) # Encuentra el Key y lo elimina
print(dic2)       
```

### popitem()

Elimina el último elemento del diccionario.

```python
dic2
```

```python
dic2.popitem()
print(dic2)
```

### update(<obj>)

Los value son actualizados y si alguna key del nuevo diccionario no esta, es añadida.

```python
dic3 = {'a': 1, 'b': 2, 'c': 3}
dic4 = {'a': [2, 5, 6, 8], 'd': 250}
```

```python
dic4.update(dic3)     # actualizamos "dic3" con "dic4"
print(dic4)
```

```python
dic3.update(dic4)     # actualizamos "dic3" con "dic4"
print(dic3)
```
