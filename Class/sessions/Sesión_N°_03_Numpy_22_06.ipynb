# Lista en Array

```python
Lista = [x*3 for x in range(1, 21)]
Lista
```

```python
# convirtiendo a Vector (i,)
lista_array = np.array(Lista)
lista_array
```

```python
# Convirtiendo a una Matriz          (i, j)
lista_array = np.array(Lista).reshape(4, 5) # F x C = nro elem Lista
lista_array
```

# Datos nulos

np.nan

```python
# creado array con data nula
a8 = np.array([5, 2, 3, np.nan, np.nan, 8, 5, np.nan])
a8
```

## Reemplazar datos nulos

array[condicin]

```python
# Incorrecta
a8[a8 == np.nan] = 10  # no genera efectos
a8
```

```python
# Correcta
a8[np.isnan(a8)] = 10  # genera efectos
a8
```

```python
a9 = np.array([  [1, 2, 3],  # Matriz con datos perdidos
                 [4, 5, 6],
                 [7, 8, 9],
                 [np.nan, 1, 2],
                 [3, 4, 5]
               ])
a9
```

```python
# identificamos datos nulos

np.isnan(a9)
```

## Seleccionando Filas sin Nulos

```python
a9[~np.isnan(a9).any(axis=1)]  # a nivel de filas -- eliminando muestras / elementos muestrales
```

```python
a9
```

## Seleccionando Columnas sin Nulos

```python
a9[:, ~np.isnan(a9).any(axis=0)] # todas las Columnas donde no se tengan NAN
```
