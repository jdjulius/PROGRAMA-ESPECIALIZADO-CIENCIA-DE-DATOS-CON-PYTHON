## Argumentos y Paramétros

1. Al definir una función los valores los cuales se reciben se denominan parámetros, pero
2. Durante la llamada los valores que se envían se denominan argumentos.

### Por posición

Cuando envía argumentos a una función, estos se reciben por orden en los parámetros definidos. Se dice por tanto que son argumentos por posición:

```python
#        0,        1
def suma(a: float, b: float)-> float:   # parámetros a y b ---> index: 0, 1
  '''
  Devuelve la suma de 2 parámetros: `2*a + b`
  y el valor de `a`.

  Parámetros
  ----------
  a : float
  b : float
  *a y b* pueden también ser números de tipo `int`.

  '''
  return 2*a + b, a
        #     0   1
```

```python
                 # argumentos
            # 0   1
x1, x2 = suma(200, 50)  # el argumento 5 ocupa la posición 0; por tanto es el parámetro a.
                 # el argumento 3 ocupa la posición 1; por tanto es el parámetro b.

print(x1)
print(x2)
```

### Por nombre

Se puede evadir el orden de los parámetros si indica durante la llamada de la función el valor tiene cada parámetro a partir de su nombre.

```python
def suma2(a:int, b, c):  #---> index a - 0 ; b - 1
  """Suma2: devuelve el cálculo de a+10*b + c*5"""
  return a + b*10 + c*5
```

```python
 # no toma en cuenta la posiciòn de nuestros paràmetros
 #               a, b, c
sum = suma2(c=100,
            b=5000,
            a=150) # ya no importa el orden de los argumentos

sum
```

### Parámetros con valores por defecto

Para solucionar la excepción **TypeError** ejecutada al momento de la llamada a una función sin argumentos, entonces se puede asignar unos valores por defecto nulos a los parámetros.

```python
def suma3(a=None, b=None):
  """Suma3: devuelve la suma de 2 parámetros: a + b*100"""

  if a == None or b == None:
    print ("Error 255!!!: enviar los argumentos a sumar")
    return           # indicamos el final de la sentencia aunque no devuelva nada

  return a + b*100       # indicamos el final de la sentencia devolviendo un cálculo

```

```python
print(suma3(50., 40.263))  #
print("-"*20)
print(suma3(b=5))
print("-"*20)
print(suma3())
```

## Argumentos indeterminados

Si no sabe previamente cuantos elementos necesita enviar a una función; en estos casos puede utilizar los parámetros indeterminados por posición y por nombre.

### Por posición

Para recibir los parámetros indeterminados por posición, definir el parámetro con un asterisco

```python
def funcion_indeterminados_posicion(*args):  # función que realiza una acción
  for i in args:
    print(i)
```

```python
#                                      0                1          2        3      4
funcion_indeterminados_posicion([1, 2, 3, 4, 5, 6], "Hola mundo", 45.6, (45, 12), 100)  # argumento <--
```

### Por nombre

Para recibir un número indeterminado de parámetros por nombre (clave-valor o en inglés keyword args), se debe crear un diccionario dinámico de argumentos definiendo el parámetro con dos asteriscos.

```python
def funcion_indeterminados_nombre(**kwargs):

  print(kwargs)
```

```python
funcion_indeterminados_nombre(lista1= [1, 2, 3, 4, 5, 6],
                              saludo= "Hola mundo",
                              num_int= 45,
                              mini_lista=[5, 4, 20],
                              m = (2, 14, 29, 100)
                              ) # diccionario dinámico
```

```python
# comprendiendo las funciones de las librerías

import pandas as pd

pd.DataFrame(data = {'col1': [1, 2],
                     'col2': [3, 4]}
             )
```

```python
# Al recibirse como un diccionario, se puede iterar y mostrar la clave y valor de cada argumento

def funcion_indeterminados_nombre(**kwargs):

  for i in kwargs:
    print (i, "=>", kwargs[i])  # Parámetro => Argumento
```

```python
funcion_indeterminados_nombre(
                              lista1= [1, 2, 3, 4, 5, 6],
                              saludo= "Hola mundo",
                              num_int= 50,
                              mini_lista=[5, 4],
                              m = (2, 14, 29)
)
```

### Por posición y nombre

Si se necesita aceptar ambos tipos de parámetros simultáneamente en una función, entonces se debe crear ambas colecciones dinámicas.

Siendo así, primero los argumentos indeterminados por valor y luego los cuales son por clave y valor.



```python
                         #  0,  1
def funcion_posicion_nombre(*args, **kwargs):  # argumentos indeterminados
  ''' funcion_posicion_nombre: es una función por posición y nombre'''
  total = 0

  for i in args:  #argumentos por posición
    total += i

  print ("Suma total:", total)

  for j in kwargs: #argumentos por nombre
    print(j, "=>", kwargs[j]) # parámetro => argumento
```

```python
#                        0     1   2     3    4     5
funcion_posicion_nombre(100, -15, 5.96, 500, 170, 5000,  Celular="977586955", Nro_hijos=2, ciudad_vive="Lima")
```

## Sentencia Return

El proceso de comunicación con el exterior se hace devolviendo valores.

```python
def calcula(a: float, b: float, TypeCal:bool=False)-> float: # paramétros

  """calcula: devuelve la suma de 2 parámetros"""
  if TypeCal == True:
    return a + b, 3*b

  return None
```

```python
sum = calcula(a=100,
              b=60) #argumentos
print(sum)
```

```python
sum, v1 = calcula(
              a=100,
              b=300,
              TypeCal=True)
print(sum)
print(v1)
```

## Función **grafico_numericos**

```python
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
def graficos_violin(columnas: list, data: pd.DataFrame, x: str): # parámetros
  '''
  Gráfico de violín de múltiples variables numéricas con una variable cualitativa.

  Parámetros
  ----------

  `columnas` : lista con los nombres de las variables numéricas.
  `data` : de tipo DataFrame.
  `x` : variable cualitativa | variable de exploración.

  Esta función graficará 1 gráfico de violín para cada variable numérica ingresada
  en la `lista`, según la variable `x`.

  Ejemplos:
  --------

  '''

  for i, colum in enumerate(columnas):
    fig, ax = plt.subplots(figsize=(6, 3.84))

    sns.boxplot(   # boxplot  violinplot
        x = x,  # eje X
        y = colum,  # eje Y
        data  = data,
        )
    ax.set_title(colum, fontsize = 10, fontweight = "bold")
    ax.tick_params(labelsize = 8)
    ax.set_xlabel("Nivel")


```

## Función **suma** con Columnas del Dataframe

```python
def suma(a:float, b:float=2500, TypeCal:bool=False)-> float:

  '''
  Retorna la suma de dos números ingresados: a + b*100.

  Parámetros
  ----------

  `a` : variable de tipo `float`.
  `b` : variable de tipo `float`, por default asume el valor de 2500.
  `TypeCal` : variable de tipo `booleana` ('True' y 'False'), por
              default es `False`.

  Devuelve la suma de ambos números si se cumple una condicional
  igual a `True`.

  Ejemplos:
  --------
  a=3
  b=5
  >>> suma(a=3, b=5, TypeCal=True) # ingresando nuestros argumentos
  >>> 503

  '''
  if TypeCal == True:
    return (a + b*100)  # retorna 1 valor
```

## Función **grafico_doble_entrada**

```python
# Etiquetamos a nuestras variables "zona y tipo de vivienda"
#-----------------------------------------------------------
data_riesgos["tipo_vivienda_r"] = data_riesgos["tipo_vivienda"].apply(lambda x:
                                                    "Vivienda A" if x==1 else (
                                                        "Vivienda B" if x==2 else (
                                                            "Vivienda C" if x==3 else "Vivienda D"
                                                        )
                                                    )
)

data_riesgos["zona_r"] = data_riesgos["zona"].apply(lambda x:
                                                    "Zona A" if x==1 else (
                                                        "Zona B" if x==2 else (
                                                            "Zona C" if x==3 else (
                                                                 "Zona D" if x==4 else "Zona E"
                                                            )
                                                        )
                                                    )
)

data_riesgos[["zona", "zona_r","tipo_vivienda", "tipo_vivienda_r"]].tail()
```

```python
ct_ztv = pd.crosstab(data_riesgos.tipo_vivienda_r, data_riesgos.zona_r, normalize='columns').round(4)*100 # 'columns'
ct_ztv
```

```python
# obtenemos las categorías
categorias_filas = list(ct_ztv.index)
categorias_filas
```

```python
# obtenemos nuestros diccionario de listas
dic1 = ct_ztv.to_dict('list')
dic1
```

```python
import numpy as np

def grafico_doble_entrada(datos: dict, nombres_filas: list):
    """
    Parámetros
    ----------

    `datos` : dict.
    `nombres_filas` : list con las nombres de las filas.

    La finalidad es graficar una tabla de doble entrada
    obtenida de una `Dataframe`.

    """
    labels = list(datos.keys())
    data = np.array(list(datos.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.get_cmap('viridis')(   # 'RdYlGn' 'viridis'
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(nombres_filas, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        ax.barh(labels, widths, left=starts, height=0.5,
                label=colname, color=color)
        xcenters = starts + widths / 2

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        for y, (x, c) in enumerate(zip(xcenters, widths)):
            ax.text(x, y, str(int(c)), ha='center', va='center',
                    color=text_color)
    ax.legend(ncol=len(nombres_filas), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='small')

    return fig, ax
```

```python
print("Gráfico de % de viviendas según zona")
print("-"*75 + "\n")
grafico_doble_entrada(dic1, categorias_filas)
plt.show()
```

# Optimizando nuestras condicionales

Creando un switch- case en Python para optimizar las condicionales

```python
def calculo_dim(calculo: str='potencia', a:float=0, b:float=0) -> float:
  ''' Función que realiza múltiples cálculos, solo debes seleccionar una operación matemática.

  Parámetros
  ----------

  `calculo` : 'suma', 'resta', 'multiplicacion', 'division', 'potencia', 'ecuación'.
  `a` : variable de tipo `float`, por default asume el valor de 0.
  `b` : variable de tipo `float`, por default asume el valor de 0.

  Ejemplos:
  --------
  >>> calculo_dim(calculo='suma')(a=10, b=30)
  >>> 40

  '''
  return {
      'suma': lambda a, b: a + b,
      'resta': lambda a, b: a - b,
      'multiplicacion': lambda a, b: a * b,
      'division': lambda a, b: a / b,
      'potencia': lambda a, b: a**b,
      'ecuacion': lambda a, b: a**2 + 5*b - 13
 }.get(calculo, lambda: "potencia")  #None
```

```python
import numpy as np
import pandas as pd

np.random.seed(1234)

dic ={
    "variable1" :  [x*2 for x in range(1, 20+1)], # datos secuenciales
    "variable2" :  list(np.random.randint(0, 50, 20)) # datos aleatorio
}

df = pd.DataFrame(dic)
df
```

```python
df["ecuacion"] = calculo_dim(calculo='ecuacion')(df.variable1, df.variable2)
df["potencia"]     = calculo_dim()(df.variable1, df.variable2)  #Por default "potencia"
df.head(7)
```
