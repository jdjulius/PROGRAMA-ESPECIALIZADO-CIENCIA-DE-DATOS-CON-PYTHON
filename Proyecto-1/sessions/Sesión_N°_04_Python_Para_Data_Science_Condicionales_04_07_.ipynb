## Sentencia "IF... "

Significa, **Si** se cumple la expresión condicional se ejecuta el bloque de sentencias seguidas.

La condición se evalúa siempre.

*   Si el resultado es True se ejecuta el bloque de sentencias.
*   Si el resultado es False no se ejecuta el bloque de sentencias.



iff.PNG

```python
a = 20
b = 3

if (b > a):           # si no se cumple la condicional, no imprimirá nada
  print("Verdadero a > b")
```

```python
numero = int(input("Ingrese un número: ")) # ingresamos

if numero < 100:                      # condición
  print("El número es menor a 100") # bloque

print(f"Ha ingresado el número {numero}") # fuera de la condicional
```

## Sentencia "IF... ELSE... "

**De lo contrario Si** se cumple la expresión condicional se ejecuta el bloque de sentencias seguidas.

La condición se evalúa siempre:


*   Si el resultado es True se ejecuta solamente el bloque de sentencias 1.
*   Si el resultado es False se ejecuta solamente el bloque de sentencias 2.






ifel.PNG

```python
edad = int(input("Ingrese su edad: "))

if edad < 18: print("Es usted menor de edad") ; print("..."); print('...')
else:
    print("Es usted mayor de edad")

print(f"Usted tiene {edad} años")  # Sentencias que estàn fuera de nuestra condicional
```

### Sentencias condicionales anidadas

Condicional dentro de otra condicional.

```python
edad = int(input("Ingrese su edad: "))

if edad < 18:

  if edad < 10:
    print (f"Tienes menos de 10 años")
  else:
    print (f"Tienes entre 10 y 17 años ")

  print("Eres menor de edad")

else:
    print("Eres mayor de edad")

print(f"Tienes {edad} años")
```

## Sentencia "IF... ELIF... ELSE"

**De lo contrario** se cumple sin evaluar ninguna expresión condicional y ejecuta el bloque de sentencias seguidas.


*   Si se cumple la condición 1, se ejecuta el bloque 1
*   Si no se cumple la condición 1 pero sí que se cumple la condición 2, se ejecuta el bloque 2
*   Si no se cumplen ni la condición 1 ni la condición 2, se ejecuta el bloque 3.

Podemos escribir tantos bloques "ELIF" como sean necesarios.
El bloque else (es opcional) se ejecuta si no se cumple ninguna de las condiciones "ELIF" anteriores.

```python
#ingreso mensual

ingreso = int(input("Ingrese su ingreso mensual (US$): "))

if ingreso >= 1000:
    print("Su ingreso es mayor igual que US$ 1000")
elif ingreso < 0:
    print("No puede tener un ingreso negativo")
else:
    print("Su ingreso es menor que US$ 1000")
```

El anterior caso fue escrito en desorden; vamos a escribir en el orden correspondiente para no perder información

```python
ingreso = int(input("Ingrese su ingreso mensual (US$): "))

if ingreso < 0:
    print("No puede tener un ingreso negativo!")
elif ingreso >= 0 and ingreso < 1000:
    print("Su ingreso es menor que US$ 1000")
else:  # >=1000
    print("Su ingreso es mayor igual que US$ 1000")
```

```python
# números múltiplos de 3 y 6

numero = int(input("Ingrese un número: "))

if numero % 3 == 0 and numero % 6 != 0:     # evaluando el residuo
    print(f"{numero} es múltiplo de 3")
elif numero % 6 == 0:
    print(f"{numero} es múltiplo de 6 y de 3")
else:
    print(f"{numero} no es múltiplo de 3")
```

Vamos a simplificar lo escrito el código anterior

```python
numero = int(input("Ingrese un número: "))

if numero % 6 == 0:                          # evaluando el residuo
    print(f"{numero} es múltiplo de 6 y de 3")
elif numero % 3 == 0:
    print(f"{numero} es múltiplo de 3")
else:
    print(f"{numero} no es múltiplo de 3 ni de 6")
```

## Operador "is"

Significa, que prueba la identidad.
Ambos lados de la expresión condicional debe ser el mismo objecto.

```python
a, b = 1, 2  # asignaciòn múltiple
a is b
```

```python
a, b = 5, 5

a is b
```

```python
5 is 5.       # fijarse en el  "." que hace la otra expresión diferente
```

```python
5 is 5
```

## Operador "in"

Significa, para cualquier *colección del valor* del lado izquierdo contenga el valor del lado derecho.

```python
lista1 = [1, 2, 3, 4, 5, 6] #creamos una lista

0 in lista1 # consultamos el valor
```

```python
5 not in lista1
```
