## Sentencia "IF... "

Significa, **Si** se cumple la expresin condicional se ejecuta el bloque de sentencias seguidas.

La condicin se evala siempre.

*   Si el resultado es True se ejecuta el bloque de sentencias.
*   Si el resultado es False no se ejecuta el bloque de sentencias.



iff.PNG

```python
a = 20
b = 3

if (b > a):           # si no se cumple la condicional, no imprimir nada
  print("Verdadero a > b")
```

```python
numero = int(input("Ingrese un nmero: ")) # ingresamos

if numero < 100:                      # condicin
  print("El nmero es menor a 100") # bloque

print(f"Ha ingresado el nmero {numero}") # fuera de la condicional
```

## Sentencia "IF... ELSE... "

**De lo contrario Si** se cumple la expresin condicional se ejecuta el bloque de sentencias seguidas.

La condicin se evala siempre:


*   Si el resultado es True se ejecuta solamente el bloque de sentencias 1.
*   Si el resultado es False se ejecuta solamente el bloque de sentencias 2.






ifel.PNG

```python
edad = int(input("Ingrese su edad: "))

if edad < 18: print("Es usted menor de edad") ; print("..."); print('...')
else:
    print("Es usted mayor de edad")

print(f"Usted tiene {edad} aos")  # Sentencias que estn fuera de nuestra condicional
```

### Sentencias condicionales anidadas

Condicional dentro de otra condicional.

```python
edad = int(input("Ingrese su edad: "))

if edad < 18:

  if edad < 10:
    print (f"Tienes menos de 10 aos")
  else:
    print (f"Tienes entre 10 y 17 aos ")

  print("Eres menor de edad")

else:
    print("Eres mayor de edad")

print(f"Tienes {edad} aos")
```

## Sentencia "IF... ELIF... ELSE"

**De lo contrario** se cumple sin evaluar ninguna expresin condicional y ejecuta el bloque de sentencias seguidas.


*   Si se cumple la condicin 1, se ejecuta el bloque 1
*   Si no se cumple la condicin 1 pero s que se cumple la condicin 2, se ejecuta el bloque 2
*   Si no se cumplen ni la condicin 1 ni la condicin 2, se ejecuta el bloque 3.

Podemos escribir tantos bloques "ELIF" como sean necesarios.
El bloque else (es opcional) se ejecuta si no se cumple ninguna de las condiciones "ELIF" anteriores.

```python
#ingreso mensual

ingreso = int(input("Ingrese su ingreso mensual (US$): "))

if ingreso >= 1000:
    print("Su ingreso es mayor igual que US$ 1000")
elif ingreso < 0:
    print("No puede tener un ingreso negativo")
else:
    print("Su ingreso es menor que US$ 1000")
```

El anterior caso fue escrito en desorden; vamos a escribir en el orden correspondiente para no perder informacin

```python
ingreso = int(input("Ingrese su ingreso mensual (US$): "))

if ingreso < 0:
    print("No puede tener un ingreso negativo!")
elif ingreso >= 0 and ingreso < 1000:
    print("Su ingreso es menor que US$ 1000")
else:  # >=1000
    print("Su ingreso es mayor igual que US$ 1000")
```
